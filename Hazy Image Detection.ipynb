```python
x1=x.drop(['Contrast Energy','Color Gradient','Saturation','Edge Detection','Covariance','Hue Disparity','Luminance','Michelson Contrast'],axis=1)
```

```python
x1.info()
```

```python
x1.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
ss.fit(x1)
```

```python
x2=pd.DataFrame(ss.transform(x1),columns=['a','b','c','d','e','f','g','h','i','j'])
```

```python
x2
```

```python
x2.corr()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
model=LinearRegression()
xtr,xte,ytr,yte=train_test_split(x2,y,test_size=0.33,random_state=1) #random state=1 means fixed
model.fit(xtr,ytr)
```

```python
model.score(xte,yte)
```

```python
from sklearn import svm
model=svm.SVC()
model.fit(xtr,ytr)
model.score(xte,yte)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(xtr,ytr)
model.score(xte,yte)
```

```python
models= {'RandomForest':RandomForestClassifier(),"SVM":svm.SVC()}
for name, model in models.items():
    model.fit(xtr,ytr)
    print(name+'trained')
```

```python
final_prediction=(0.6*(models['RandomForest'].predict(xte))+0.4*(models['SVM'].predict(xte)))
```

```python
final_prediction
```

```python
#r2 square to predict the model score
from sklearn.metrics import r2_score 
r2_score(yte,final_prediction)
```
