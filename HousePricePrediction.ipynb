```python
#remove outliers
```

```python
q1=df['Sale Price'].quantile(0.25)
q3=df['Sale Price'].quantile(0.75)
```

```python
iqr=q3-q1
```

```python
iqr
```

```python
upper_limit=q3+1.5*iqr
lower_limit=q1-1.5*iqr
upper_limit,lower_limit
```

```python
def limit_inputer(value):
    if value>upper_limit:
        return upper_limit
    elif value<lower_limit:
        return lower_limit
    else:
        return value

    
```

```python
df['Sale Price']=df['Sale Price'].apply(limit_inputer)
```

# Treating missing values

```python
#treating missing values- row wise deltion for the target variable
# deletion and imputing(in imputing we replace missing data with some values- mean median for the continous variable and mode for the object variables)
```

```python
#first from the independent variable
df.dropna(inplace=True,axis=0,subset=['Sale Price'])
```

```python
df.info()
```

```python
# from the depandent variables
"""import the SimpleImputer from the sklearn.impute and the impute the missing values"""
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='median')
```

```python
numerical_column=['No of Bathrooms','Flat Area (in Sqft)','Lot Area (in Sqft)','Area of the House from Basement (in Sqft)','Zipcode','Latitude','Longitude','Living Area after Renovation (in Sqft)']
```

```python
df[numerical_column]=imputer.fit_transform(df[numerical_column])
```

```python
df.info()
```

```python
df["Zipcode"]=df['Zipcode'].astype(object)
```

```python
df.info()
```

```python
mapping={'None':'0',
         'Once':'1',
         'Twice':'2',
         'Thrice':'3',
         'Four':'4'
}
df['No of Times Visited']=df['No of Times Visited'].map(mapping)
```

```python
df['Year Since Renovation']=np.where(df['Ever Renovated']=='Yes',
                                    abs(df['Purchase Year']-df['Renovated Year']),0)
```

# Binning

```python
Zip_Table=df.groupby('Zipcode').agg({'Sale_Price':'mean'}).sort_values('Sale_Price',ascending=True)
```

```python
Zip_Table['Zipcode_Group']=pd.cut(Zip_Table['Sale_Price'],bins=10,labels=['Zipcode_Group1','Zipcode_Group2','Zipcode_Group3',
                                                                         'Zipcode_Group4','Zipcode_Group5','Zipcode_Group6',
                                                                         'Zipcode_Group7','Zipcode_Group8','Zipcode_Group9','Zipcode_Group10'
                                                                         ],include_lowest=True)
```

```python
Zip_Table.head()
```

```python
Zip_Table=Zip_Table.drop(columns=['Sale_Price'])
```

```python
Zip_Table.head()
```

```python
df=pd.merge(df,Zip_Table,left_on='Zipcode',how='left',right_index=True)
```

```python
df.head()
```

```python
df=pd.get_dummies(df,columns=['Zipcode_Group'],drop_first=True)
```

```python
df.head()
```

# Spilliting of data

```python
Y=df.iloc[:,0]
```

```python
X=df.iloc[:,1:]
```

```python
# now we are going to split the data into train and test data
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtr,xte,ytr,yte=train_test_split(X,Y,test_size=0.30,random_state=1)
```

# Feature Scaling

```python
# three common methods are- standardization, min max scaling, normalization
#here in this case we are using standardisation method
```

```python
from sklearn import preprocessing
scale=preprocessing.StandardScaler()
```

```python
xtr=scale.fit_transform(xtr)
xte=scale.fit_transform(xte)
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(xtr, ytr)
model.score(xte,yte)
```
